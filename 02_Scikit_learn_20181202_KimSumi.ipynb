**1. Libraries**

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

**2. Load Dataset**

```python
# Load the diabetes dataset
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
# Use only one feature
diabetes_X = diabetes_X[:, np.newaxis, 2]
```

```python
graph = plt.scatter(diabetes_X, diabetes_y, label='Dataset')
plt.legend()
plt.show(graph)
```

**3. Split data : Training & Testing**

```python
# Split the data into training/testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

# Split the targets into training/testing sets
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]
```

```python
# Train
graph = plt.scatter(diabetes_X_train, diabetes_y_train, label='Train')
plt.legend()
plt.show(graph)
```

```python
# Test
graph = plt.scatter(diabetes_X_test, diabetes_y_test, label='Test')
plt.legend()
plt.show(graph)
```

**4. Create Model**

```python
# Create linear regression object
regr = linear_model.LinearRegression()
```

**5. Train Model**

```python
# Train the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)

# Make predictions using the testing set
diabetes_y_pred = regr.predict(diabetes_X_test)
```

**6. Show Results**

```python
# The coefficients
print('Coefficients: ', regr.coef_)
# The mean squared error
print('Mean squared error: %.2f' % mean_squared_error(diabetes_y_test, diabetes_y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f' % r2_score(diabetes_y_test, diabetes_y_pred))
```

**7. Plot Results**

```python
# Plot outputs with test set
plt.scatter(diabetes_X_test, diabetes_y_test, label='Test')
plt.plot(diabetes_X_test, diabetes_y_pred, color = 'red', linewidth=2, label='Predicted')

plt.xticks(())
plt.yticks(())
plt.legend()
plt.show()
```

```python
# Plot outputs with train set
diabetes_y_train_pred = regr.predict(diabetes_X_train)
plt.scatter(diabetes_X_train, diabetes_y_train, label='Train')
plt.plot(diabetes_X_train, diabetes_y_train_pred, color='red', linewidth=2, label='Predicted')

plt.xticks(())
plt.yticks(())
plt.legend()
plt.show()
```

**8. Training Process**

```python
# Regression Model
regr = linear_model.LinearRegression()

# Process
regr.fit(diabetes_X_train[-2:], diabetes_y_train[-2:])
diabetes_y_pred1 = regr.predict(diabetes_X_test)
diabetes_y_train_pred1 = regr.predict(diabetes_X_train)

regr.fit(diabetes_X_train[-5:], diabetes_y_train[-5:])
diabetes_y_pred2 = regr.predict(diabetes_X_test)
diabetes_y_train_pred2 = regr.predict(diabetes_X_train)

regr.fit(diabetes_X_train[-8:], diabetes_y_train[-8:])
diabetes_y_pred3 = regr.predict(diabetes_X_test)
diabetes_y_train_pred3 = regr.predict(diabetes_X_train)

regr.fit(diabetes_X_train, diabetes_y_train)
diabetes_y_pred4 = regr.predict(diabetes_X_test)
diabetes_y_train_pred4 = regr.predict(diabetes_X_train)
```

```python
# Plot : Test
plt.scatter(diabetes_X_test, diabetes_y_test, color = 'darkgrey', label='Test')
plt.plot(diabetes_X_test, diabetes_y_pred1, color = 'deepskyblue', linewidth=2, label='Pred1')
plt.plot(diabetes_X_test, diabetes_y_pred2, color = 'dodgerblue', linewidth=2, label='Pred2')
plt.plot(diabetes_X_test, diabetes_y_pred3, color = 'blue', linewidth=2, label='Pred3')
plt.plot(diabetes_X_test, diabetes_y_pred4, color = 'red', linewidth=2, label='Pred4')

plt.xticks(())
plt.yticks(())
plt.legend()
plt.show()
```

```python
# Plot : Train
plt.scatter(diabetes_X_train, diabetes_y_train, color = 'darkgrey', label='Train')
plt.plot(diabetes_X_train, diabetes_y_train_pred1, color='deepskyblue', linewidth=2, label='Pred1')
plt.plot(diabetes_X_train, diabetes_y_train_pred2, color = 'dodgerblue', linewidth=2, label='Pred2')
plt.plot(diabetes_X_train, diabetes_y_train_pred3, color = 'blue', linewidth=2, label='Pred3')
plt.plot(diabetes_X_train, diabetes_y_train_pred4, color = 'red', linewidth=2, label='Pred4')

plt.xticks(())
plt.yticks(())
plt.legend()
plt.show()
```
